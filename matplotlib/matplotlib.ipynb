```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as py
```

```python
gas = pd.read_csv("gas_prices.csv")
gas
```

```python
plt.figure(figsize=(8,5))

plt.plot(gas["Year"], gas["France"], "b.-")
plt.plot(gas["Year"], gas["UK"], "g.-")

plt.xticks(gas["Year"][::3])
plt.legend(["France", "UK"])


plt.xlabel("Years")
plt.ylabel("USD $")
plt.grid()


plt.show()
```
